# MERGING THE DATA

```python
cluster=pd.merge(df_district,df_sctot,how='inner',left_on=['STATE/UT','DISTRICT','YEAR'],right_on=['STATE/UT','DISTRICT','Year']).merge(df_sttot,how='inner',left_on=['STATE/UT','DISTRICT','Year'],right_on=['STATE/UT','DISTRICT','Year']).merge(df_childtot,how='inner',left_on=['STATE/UT','DISTRICT','Year'],right_on=['STATE/UT','DISTRICT','Year']).merge(df_womentot,how='inner',left_on=['STATE/UT','DISTRICT','Year'],right_on=['STATE/UT','DISTRICT','Year'])
```

# GROUPING BY STATE/UT,DISTRICT AND SUMMING ALL THE CRIMES OVER THE YEARS



```python
cluster=cluster.groupby(['STATE/UT','DISTRICT']).sum().reset_index().drop(['Year'],axis=1)
```

```python
cluster
```

# 4.1  "Create 3 clusters as below.



```python
x=cluster.iloc[:,2:]
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans(n_clusters=3,random_state=0)
```

```python
y_kmeans=kmeans.fit_predict(x)
```

# 1. Sensitive Area's



```python
peaceful_regions=cluster[y_kmeans==0]
```

```python
peaceful_regions
```

# 2. Moderate Area's



```python
moderate_regions=cluster[y_kmeans==1]
```

```python
moderate_regions
```

# 3. Peaceful Area's"



```python
sensitive_regions=cluster[y_kmeans==2]
```

```python
sensitive_regions
```

# 5 MOST SENSITIVE AREAS


```python
sensitive_regions['OVERALL CRIME']=sensitive_regions.iloc[:,2:].sum(axis=1)
```

```python
sensitive_regions.sort_values(by='OVERALL CRIME',ascending=False,inplace=True)
```

```python
sensitive=sensitive_regions.reset_index().drop(['index'],axis=1).head(5)
```

```python
sensitive
```

```python
plt.figure(figsize=(8,8),facecolor='Pink')
sns.barplot(x=sensitive['DISTRICT'],y=sensitive['OVERALL CRIME'])
plt.xticks(rotation=90)
plt.xlabel("DIFFERENT STATES/UTs")
plt.title("TEN MOST SENSITIVE AREAs(DISTRICTS)!!")
plt.grid()
plt.show()
```
